#### Explore site geographical data

```python
## Check the geographical data

locations = site_metadata[['latitude','longitude']].dropna()

locationlist = locations.values.tolist()


map = folium.Map(location=[41, -89], zoom_start=7)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=f'Site displayed #{point}').add_to(map)

folium.Marker(locationlist[point], popup=f'Site displayed #{point}', icon=folium.Icon(color='red')).add_to(map)

display(map)

print(f'Only {len(locations)} can be displayed on the Map as 22 sites of USGS_OFR do not contain known lat long')
```

```python
def print_df_specs(df):
    print(df.filename[0])
    print(df.dtypes)
    print('\n')
    print(df.describe())
    print('\n Any duplicates: \n')
    print(df[df.duplicated()])
    print('\n')

counter = 1
for df in d_m_df:
    print('data file', counter)
    ## run df specs function
    print_df_specs(df)
    counter = counter + 1
    
```

### Dates Check 

```python
for z in d_m_df:
    
    filename = z['filename'].iloc[0]
    
    print('\n', filename)
    
    print(z.iloc[:,0:5].head())
    
    if (z.columns.str.contains('date').any() == True):
        z.sort_values('date')
        functions_notebook.check_date(z, date_col = 'date')
    else:
        print(f'No date column in {filename}')
        print('columns: ', z.columns)
```

### Tabular Values check

```python

counter = 1
for df_1 in d_m_df:
    print(f'\n{counter}. {df_1.filename[0]}')
    cols = df_1.columns
    for col in cols:
        print('   ', df_1[col].name,'  - dtype:', df_1[col].dtypes)
        if df_1[col].dtypes == 'O':
            functions_notebook.check_whitespace(df_col = df_1[col])
        if (df_1[col].dtypes == 'float64') or (df_1[col].dtypes == 'int64'):
            functions_notebook.calc_metrics(df_col = df_1[col])
        else: 
            pass
    counter = counter + 1
    print(f'\n   > whitespace scan complete for {df_1.filename[0]}')

        
```

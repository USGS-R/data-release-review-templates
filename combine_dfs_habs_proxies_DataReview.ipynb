### Dates Check

```python
counter = 1
for z in comb_df:
    
    print(z.iloc[:,0:5].head())
    
    filename = z['filename'].iloc[0]
    
    print('\n',counter,'.', filename)

    if (z.columns.str.contains('date').any() == True):
        z.sort_values('date')
        functions_notebook.check_date(z, date_col = 'date')
    else:
        print(f'No date column in {filename}')
        print('columns: ', z.columns)
        
    counter = counter + 1
```

### Tabular Values

```python

counter = 1
for df in comb_df:
    print(f'\n{counter}. {df.filename[0]}')
    cols = df.columns
    for col in cols:
        print('   ', df[col].name,'  - dtype:', df[col].dtypes)
        if df[col].dtypes == 'O':
            functions_notebook.check_whitespace(df_col = df[col])
        if (df[col].dtypes == 'float64') or (df[col].dtypes == 'int64'):
            functions_notebook.calc_metrics(df_col = df[col])
        else: 
            pass
    counter = counter + 1
    print(f'\n   > whitespace scan complete for {df.filename[0]}')

        
```
